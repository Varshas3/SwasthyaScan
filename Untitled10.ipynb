```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory

# Dataset parameters
DATA_DIR = "dataset"   # adjust to your dataset folder
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# Load training and validation sets
train_ds = image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_ds = image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)



# Normalize AFTER capturing classes
# AFTER вЂ” move the import to the top of cell 3, or merge cells:
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

train_ds = image_dataset_from_directory(...)
val_ds   = image_dataset_from_directory(...)
classes  = train_ds.class_names
print("Classes:", classes)

train_ds = train_ds.map(lambda x, y: (preprocess_input(x), y))
val_ds   = val_ds.map(lambda x, y: (preprocess_input(x), y))
```

```python
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Data augmentation
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.15),
    layers.RandomZoom(0.2),
    layers.RandomContrast(0.1),
])

# Base model
base_model = MobileNetV2(input_shape=IMG_SIZE + (3,), include_top=False, weights="imagenet")
base_model.trainable = True

# Freeze all but last 40 layers
for layer in base_model.layers[:-40]:
    layer.trainable = False

# Classification head
model = models.Sequential([
    data_augmentation,
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation="relu", kernel_regularizer=regularizers.l2(0.01)),
    layers.Dropout(0.5),
    layers.Dense(len(classes), activation="softmax")
])

# Compile
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

# Callbacks
early_stop = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.2, patience=3, min_lr=1e-6)

# Train
history = model.fit(train_ds,
                    validation_data=val_ds,
                    epochs=30,
                    callbacks=[early_stop, reduce_lr])

# Save model
model.save("health_model.keras")
print("Model saved as health_model.keras")
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load trained model
model = load_model("health_model.keras")
print("Classes:", classes)

def preprocess_image(img_path):
    img = cv2.imread(img_path)
    if img is None:
        raise FileNotFoundError(f"Image not found: {img_path}")
    img = cv2.resize(img, (224, 224))
    img = img / 255.0
    return img.astype(np.float32)

# Test multiple images
test_image_paths = []

images, valid_paths = [], []
for path in test_image_paths:
    if os.path.exists(path):
        images.append(preprocess_image(path))
        valid_paths.append(path)
    else:
        print("File does not exist:", path)

if images:
    batch = np.array(images, dtype=np.float32)
    predictions = model.predict(batch)

    for i, pred in enumerate(predictions):
        predicted_class = classes[np.argmax(pred)]
        print(f"\nImage: {valid_paths[i]}")
        print(f"Detected deficiency в†’ {predicted_class}")
        for idx, prob in enumerate(pred):
            print(f"   {classes[idx]}: {prob:.2f}")
```
